```python
#packages/labraries

import os #provides functions for interacting with the operator
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#1 Loading the raw Data
```

```python
#loading the data
raw_data = pd.read_csv("Marketing Raw Data.csv")
#run all data
raw_data
print(raw_data.shape)

#runs the first 5 rows
raw_data.head()
```

```python
#2 Line Graph
```

```python
#Example 1 - Simple 1 line graph
#Assuming we want to investigate the revenue by date
sns.lineplot(x='Week_ID',y='Revenue',data = raw_data,ci=False)
```

```python
raw_data.Promo.value_counts()
```

```python
#Example2 - Adding Categories

sns.set(rc={'figure.figsize':(6,4)})

#by Promo
sns.lineplot(x='Week_ID', y='Revenue', hue='Promo', style = 'Promo', data= raw_data,ci=False)
```

```python
#Exa,ple 4 - By Promo with style & increase the size & remove


#increase the size
sns.set(rc={'figure.figsize':(7,5)})
ax = sns.lineplot(x='Week_ID',y='Revenue', hue = 'Promo', style ='Promo', data = raw_data, ci=None)
```

```python
#Example5 by promo with style and increase the size & remove error bar& adding markers & by month

ax=sns.lineplot(x='Month', y='Revenue', hue= 'Promo',style ="Promo", data= raw_data, ci=None, markers=True)
```

```python
#Example 6

ax=sns.lineplot(x='Month', y='Revenue', hue= 'Day_Name',style ="Promo", data= raw_data, ci=None, markers=True)
```

```python
#Example 7

ax=sns.lineplot(x='Year', y='Revenue', hue= 'Day_Name',style ="Promo", data= raw_data, ci=None, markers=True)
```

```python
#Bar Plots
```

```python
#Example1 - Total Revenue by Month

ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data)
raw_data[['Month_ID', 'Revenue']]
.groupby('Month_ID', as_index = False)
.agg({'Revenue':'mean'})
```

```python
#Example2 

ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data,ci=False)
```

```python
#Example3

ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data,ci=False, hue = 'Promo')
```

```python
#Example4 

ax = sns.barplot(x="Revenue", y="Year", ci=False, hue = 'Promo', orient = 'h',data=raw_data)
```

```python
#Example5 

ax = sns.barplot(x="Revenue", y="Year", ci=False, hue = 'Promo', orient = 'h',data=raw_data, color="#1CB3B1")
```

```python
#Histograms
```

```python
#Example1 - Investigating the distribution of revenue

x = raw_data['Revenue'].values
sns.distplot(x, color = 'blue');
```

```python
#Example2

x = raw_data['Revenue'].values
sns.distplot(x, color = 'blue');
#calculating the mean
mean = raw_data['Revenue'].mean()

#plottingthe mean
plt.axvline(mean, 0,1, color = 'red')
```

```python
#Example3


x = raw_data['Visitors'].values
sns.distplot(x, color = 'red');

mean = raw_data['Visitors'].mean()

plt.axvline(mean, 0,1, color = 'blue')
```
