```python
#To add the markers:
plt.plot(a,b,"ro")
plt.title('Squared value plot')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#To add the ro-- for markers with dotted lines
plt.plot(a,b,"ro--")
plt.title('Squared value plot')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#To Plot the 2 graps in one plotting:
c=[1,8,27,64]
plt.plot(a,b,"ro--",a,c,"bx--")
plt.title('Squared value plot')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#3 Graphs in one squared_plot:
c=[1,8,27,64]
plt.plot(a,b,"ro--",a,c,"bx--",a,b*10,"g^--")
plt.title('Squared value plot')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#For in One Frame how to plot 2 plots:
plt.subplot(1,2,1)
plt.plot(a,b,"ro")
plt.title("1st plot")

plt.subplot(1,2,2)
plt.plot(a,b,"g^")
plt.title("2nd plot")

plt.suptitle("Subplot examples")
```

```python
#Plot the 2 Subplot Horizontally in one Plot:
plt.subplot(2,1,1)
plt.plot(a,b,"ro")
plt.title("1st plot")

plt.subplot(2,1,2)
plt.plot(a,b,"g^")
plt.title("2nd plot")

plt.suptitle("Subplot examples")

```

```python
#Plotting of 4 subplot in one Horizontal Frame:
plt.subplot(2,2,1)
plt.plot(a,b,"ro")
plt.title("1st plot")

plt.subplot(2,2,2)
plt.plot(a,b,"g^")
plt.title("2nd plot")


plt.subplot(2,2,3)
plt.plot(a,b,"ro")
plt.title("3rd plot")

plt.subplot(2,2,4)
plt.plot(a,b,"g^")
plt.title("4th plot")
plt.suptitle("Subplot examples")
plt.tight_layout()
```

```python
division=['A','B','C','D']
avg_mark=[55,66,77,88]

plt.bar(division,avg_mark,color='red')
plt.title('Marks ddistribution')
plt.xlabel('Division')
plt.ylabel('avg_marks')
```
